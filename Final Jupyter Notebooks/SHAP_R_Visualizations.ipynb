```r
library(ggplot2)
library(readr)
library(gridExtra)
library(extrafont)  # Ensure the extrafont library is loaded

# Note: before running plots, please add all "SVM SHAP plots" in the "svm shaps" folder and all "Extended Feature Set SHAP plots" in the "efs shaps"
create_plot <- function(file_path) {
  # Extract the string between the last two slashes
  filename <- basename(file_path)
  # Extract the name by removing "treatment" and ".csv"
  name <- gsub("\\_svm.csv$", "", filename)          # Remove ".csv"
  # Capitalize the first letter of each word
  treatment <- tools::toTitleCase(name)
  data <- read.csv(file_path)
  data$Feature <- as.factor(data$Feature)
  ggplot(data, aes(y = `Average.SHAP.Value`, x = reorder(`Feature`, `Average.SHAP.Value`), group = `Feature`)) +
      geom_bar(stat = "identity", fill = "steelblue", width = 0.4, position = position_dodge(width = 0.1)) +
      coord_flip() +
      labs(x = "", y = "") +
      ggtitle(treatment) +
      theme_minimal(base_family = "Helvetica") +  # Set the base font family to Helvetica
      theme(panel.grid.major = element_blank(),
            panel.grid.minor = element_blank(),
            plot.margin = unit(c(0.5,0.5,0.5,0.5), "cm"))
}

# Set dimensions to match Jupyter notebook rendering
options(repr.plot.width=15, repr.plot.height=2)

# Define the folder containing the CSV files
folder_path <- "/content/base shaps"
csv_files <- list.files(folder_path, pattern = "\\.csv$", full.names = TRUE)

# Create plots from CSV files
plots <- lapply(csv_files, create_plot)

# Arrange plots in a grid layout
g <- grid.arrange(grobs = plots, ncol = 2)

# Save the plot ensuring dimensions match what is rendered in Jupyter
ggsave(filename = "2_subplot.jpg", plot = g, dpi = 650, width = 15, height = 2, units = "in")
```
